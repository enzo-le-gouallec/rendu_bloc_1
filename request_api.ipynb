```python
import requests

url = "http://127.0.0.1:8000/register/"
data = {
    "email": "john@example.com",
    "name": "Doe",
    "firstname": "John",
    "password": "mypassword"
}

response = requests.post(url, json=data)
print(response.json())

```

```python
import requests

url = "http://127.0.0.1:8000/login/"
data = {
    "email": "john@example.com",
    "password": "mypassword"
}

response = requests.post(url, json=data)
token = response.json().get("token")
print(response.json())  # Affiche le token JWT
```
